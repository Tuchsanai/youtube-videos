### Let's see how to get the entire 6 years worth of data|

```python
import requests
from datetime import datetime, timedelta
symbol='TSLA'
# API endpoint
url = 'https://data.alpaca.markets/v1beta1/news'

# Set the date range for the last 6 years
end_date = datetime.now()
 
end_date_str = end_date.strftime('%Y-%m-%d')

# Function to fetch data with pagination
def fetch_data(next_page_token=None):
    params = {
        'symbols':symbol,
        'end': end_date_str,
        'limit': 50,
        'include_content': True,
        'page_token': next_page_token  # Add next_page_token to params
    }

    # Headers with API credentials
    headers = {
        'Apca-Api-Key-Id': api_key,
        'Apca-Api-Secret-Key': api_secret
    }

    # Make the request and collect the data
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print('Failed to fetch data: ', response.status_code)
        return None

# Main loop for pagination
data = []
next_page_token = None
while True:
    response_data = fetch_data(next_page_token)
    if response_data:
        data.extend(response_data.get('news', []))
        next_page_token = response_data.get('next_page_token')
        if not next_page_token:
            break
    else:
        break
    print("CREATED_AT",data[-1]['created_at'])

# Process the data as needed
# For example, print the number of news items fetched
print(f"Total news items fetched: {len(data)}")
```

### Save results to CSV FILE

```python
import pandas as pd
pd.DataFrame(data).to_csv(f"""{symbol}_news.csv""",index=False)
```
